## Q4 : In the suburb of Abbotsford, what are the chances of finding a property with 3 rooms? Round your answer to 3 decimal places. 

```python
abbotsford_data = df[df['Suburb'] == 'Abbotsford']# Filter data for Abbotsford

total_properties = len(abbotsford_data)# Count total properties in Abbotsford

properties_with_3_rooms = len(abbotsford_data[abbotsford_data['Rooms'] == 3]) # Count properties with 3 rooms

probability = properties_with_3_rooms / total_properties # Calculate the probability

#probability_rounded = round(probability, 3)# Round off the result to 3 decimal places
print(f"Total properties in Abbotsford: {total_properties}")
print(f"Properties with 3 Rooms: {properties_with_3_rooms}")
print(f"Probability of finding a property with 3 rooms: {probability:.3f}") # showing and Rounding answer to 3 decimal places
```

**Conclusion:**
- Among all properties in Abbotsford (56 in total), a significant portion (20 out of 56) consists of 3-room properties. This makes 3-room properties relatively common.
- If you randomly select a property in Abbotsford, the chance of it having 3 rooms is approximately 35.7%.


## Q5 : In the suburb of Abbotsford, what are the chances of finding a property with 2 bathrooms? Round your answer to 3 decimal places.

```python
abbotsford_data = df[df['Suburb'] == 'Abbotsford'] # Filter data for Abbotsford

total_properties = len(abbotsford_data) # Count total properties in Abbotsford

properties_with_2_bathrooms = len(abbotsford_data[abbotsford_data['Bathroom'] == 2]) # Count properties with 2 bathrooms

probability = properties_with_2_bathrooms / total_properties # Calculate the probability
print(f"Total properties in Abbotsford: {total_properties}")
print(f"Properties with 2 bathrooms: {properties_with_2_bathrooms}")
print(f"Probability of finding a property with 2 bathrooms: {probability:.3f}") # showing and Rounding answer to 3 decimal places
```

**Conclusion:**
- Among all properties (56 in total), a considerable portion (19 out of 56) has 2 bathrooms, making them relatively common but slightly less frequent than 3-room properties.
- If you randomly select a property in Abbotsford, the chance of it having 2 bathrooms is 33.9%.
